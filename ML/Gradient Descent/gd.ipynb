```python
def gradient(x):
    return 2 * (x + 3)
```

```python
def function(x):
    return (x + 3) ** 2
```

```python
def gradient_descent(start_x, lr, iters):
    x_values = [start_x]
    x = start_x
    
    for i in range(iters):
        grad = gradient(x)
        x = x -lr * grad
        x_values.append(x)
      
    return x_values
```

```python
start_point = 2
lr = 0.2
num_iters = 100

steps = gradient_descent(start_point, lr, num_iters)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y = function(x)

plt.plot(x , y)
plt.scatter(steps, function(np.array(steps)), c="red")
```
